```python
# numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
for i,col in enumerate(numeric_columns,1):
    print(f'{i}. {col:<15} | Min : {df[col].min()} | Max : {df[col].max()}')
```

```python
# categorical columns
categorical_columns = df.select_dtypes(exclude=[np.number]).columns
for i,col in enumerate(categorical_columns,1):
    uniques = df[col].unique()
    print(f'{i}. {col:<20} | Unique : {df[col].nunique():<7} | Examples : {uniques[:4]}')
```

```python
# missing values
missing_values = df.isnull().sum()
missing_pct = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({
    'missing' : missing_values,
    'misssing_pct' : missing_pct.round(2)
}).sort_values(by='missing', ascending=False)
missing_df
```

```python
# duplicated values
df[df.duplicated()]
```

```python
# outliers
for i, col in enumerate(numeric_columns,1):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f'{i}. {col:<15} | Outliers : {len(outliers):<3} | Range : ({lower_bound} - {upper_bound})')
```
